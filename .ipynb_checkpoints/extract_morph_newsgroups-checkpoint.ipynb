```python
from sklearn.datasets import fetch_20newsgroups
import numpy as np
```

```python
import elasticsearch as es
hes = es.Elasticsearch([
    {'host': '59.10.222.69', 'port': 19210},
])
```

```python
dataset = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
docs = dataset['data']
```

```python
len(docs)
```

```python
MIN_NUM_TOKENS = 15
texts = []
for dInd, d in enumerate(docs[:]):
    qb = {
        "analyzer": "rebuilt_english",
        "text": d
    }
    tokensDic = hes.indices.analyze(index='l2v_analyzer_index2', body=qb)
    tokens = [i['token'] for i in tokensDic['tokens']]
    if len(tokens) < MIN_NUM_TOKENS:
        print(f'dInd{dInd}: MIN NUM tokens is {MIN_NUM_TOKENS}')
        continue
    texts.append((dInd, ' '.join(tokens)))
    
```

```python
len(texts)
```

```python
import json
with open('./data/newsgroups_texts.json', 'w') as fp:
    json.dump(texts, fp)
```
